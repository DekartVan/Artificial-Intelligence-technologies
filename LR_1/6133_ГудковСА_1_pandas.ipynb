```python
# Чтобы продолжить работу - необходимо внести изменения в исходные данные. Нужно убрать символ "k" после чисел.  
# Список столбцов, которые требуют корректировки
columns_to_correct = ['Total_reviews', 'Avg_salary', 'Interviews_taken', 'Total_jobs_available', 'Total_benefits']

# Проходим по каждому столбцу из списка
for column in columns_to_correct:
    # Преобразуем числовые значения в строковый тип данных
    df[column] = df[column].astype(str)
    # Заменяем некорректные значения на NaN
    df[column] = df[column].replace('--', float('nan'))
    # Применяем замену символа "k" и преобразование в числовой тип данных
    df[column] = df[column].replace('k', '', regex=True).astype(float)
```

## 4. Продемонстрировать работу .groupby, на основе группировок в groupby вычислить агрегатные функции по одной или нескольким колонкам

```python
# Группируем данные по столбцу 'Ratings' и вычисляем агрегатные функции для 'Avg_salary':
grouped = df.groupby(['Ratings'])[['Avg_salary', 'Total_reviews', 'Total_benefits']].mean()
grouped
```

## 7. Вычислить квантили распределения какого-либо вещественного признака (с использованием numpy.quantile или numpy.percentile).

```python
# Удаляем строки с пропущенными значениями в столбце "Avg_salary"
cleaned_data = df['Avg_salary'].dropna()

# Вычисляем 25-й, 50-й и 75-й процентили для очищенных данных
quantiles = np.percentile(cleaned_data, [25, 50, 75])

# Выводим результаты
print("25-й процентиль:", quantiles[0])
print("50-й процентиль:", quantiles[1])
print("75-й процентиль:", quantiles[2])
```

## 8. Вычислить (в виде текста) гистограмму какого-либо вещественного признака (с использованием numpy.histogram). Значения гистограммы можно использовать в качестве загрубленного числового признака для заданий (5) или (6)

```python
# Вычисляем гистограмму для столбца "Total_reviews"
hist, bins = np.histogram(cleaned_data, bins=5)  # разбиваем на 5 интервалов

# Выводим гистограмму в виде текста
print("Гистограмма для столбца 'Avg_salary':")
for i in range(len(hist)):
    print(f"{bins[i]} - {bins[i+1]}: {hist[i]}")
```

## 10. Получить DataFrame с MultiIndex любым способом: через конструктор (в документации есть множество видов конструкторов для создания MultiIndex с нуля), через read_sql / read_csv / read_excel, read_*, через pivot_table, через groupby или иными способами.

```python
# Создаем DataFrame с MultiIndex
arrays = [
    ['A', 'A', 'B', 'B'],
    ['X', 'Y', 'X', 'Y']
]
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=['First', 'Second'])
df_multiindex = pd.DataFrame([[1, 2], [3, 4], [5, 6], [7, 8]], index=index, columns=['Value1', 'Value2'])

# Выведем DataFrame с MultiIndex
print("DataFrame с MultiIndex:")
print(df_multiindex)

# Переставим уровни индекса
df_multiindex_swapped = df_multiindex.swaplevel('First', 'Second')
print("\nПереставленные уровни индекса:")
print(df_multiindex_swapped)

# Транспонируем таблицу
df_transposed = df_multiindex.T
print("\nТранспонированная таблица:")
print(df_transposed)

# Добавим новый уровень индекса
df_with_new_index_level = df_multiindex.set_index('Value1', append=True)
print("\nDataFrame с новым уровнем индекса:")
print(df_with_new_index_level)
```

## 11. Продемонстировать работу .merge.

```python
# Создадим два DataFrame для примера
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

# Пример объединения по ключу
merged_df = pd.merge(df1, df2, on='key')
print("Объединенный DataFrame:")
merged_df
```

## 12. Продемонстрировать работу с .concat или append.

```python
# Создадим два DataFrame для примера
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7']
})

# Пример использования метода .concat() для объединения по строкам
concatenated_rows = pd.concat([df1, df2])
print("Объединение по строкам:")
concatenated_rows

```

```python
# Пример использования метода .concat() для объединения по столбцам
concatenated_columns = pd.concat([df1, df2], axis=1)
print("\nОбъединение по столбцам: ")
concatenated_columns
```
