# 2. Бинарная классификация

### 2.1 Привести описание датасета

Приведу описание датасета "Pima Indians Diabetes Database" - https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database

Датасет "Pima Indians Diabetes Database" содержит данные, предоставленные Национальным институтом диабета и болезней почек. Цель датасета - диагностическое предсказание наличия диабета у пациентов на основе определенных диагностических измерений, включенных в датасет. Датасет состоит из нескольких медицинских предикторов (независимых переменных) и одной целевой переменной, "Outcome". Предикторы включают в себя количество беременностей, уровень глюкозы в плазме крови, артериальное давление, толщину складки кожи, уровень инсулина, индекс массы тела, генетическую предрасположенность к диабету, а также возраст пациента. Целевая переменная "Outcome" указывает на наличие или отсутствие диабета (1 - присутствует, 0 - отсутствует).

| Признак              | Описание                                                                           | Тип                    |
|----------------------|------------------------------------------------------------------------------------|-----------------------|
| Pregnancies          | Число беременностей у пациента                                                      |   Количественный       |
| Glucose              | Концентрация глюкозы в плазме крови через 2 часа после приема глюкозы              |   Количественный       |
| BloodPressure        | Диастолическое давление (мм рт.ст.)                                                 |   Количественный       |
| SkinThickness        | Толщина складки кожи на трицепсе (мм)                                               |   Количественный       |
| Insulin              | Уровень инсулина в крови через 2 часа после приема глюкозы (мкед/мл)               |   Количественный       |
| BMI                  | Индекс массы тела (вес в кг / (рост в м)^2)                                          |   Количественный       |
| DiabetesPedigreeFunction | Генетическая предрасположенность к диабету (производная от истории заболевания у родственников) | Количественный |
| Age                  | Возраст пациента (лет)                                                              |   Количественный       |
| Outcome	           | Наличие диабета (0 - отсутствует, 1 - присутствует)	                             | Категориальный |ориальный

### 2.4 Выбор гиперпораметра k

```python
k = 7
```

### 2.5 Разбить выборку на обучающую и тестовую

```python
from sklearn.model_selection import train_test_split

X = df.drop('Outcome', axis = 1)
y = df['Outcome'] 

X_train, X_test, y_train, y_test = train_test_split(
                X, y, test_size=0.33, random_state=42)
```

### 2.6 Осуществить бинарную классификацию

```python
neigh = KNeighborsClassifier(n_neighbors = k)
neigh.fit(X_train, y_train)

predict = neigh.predict(X_test)
```

### 2.7 Посчитать метрики: accuracy, precision, recall, f-measure, а также составить classification_report

```python
print(f'accurasy:\t{accuracy_score(y_test, predict)}')
print(f'precision:\t{precision_score(y_test, predict)}')
print(f'recall:\t\t{recall_score(y_test, predict)}')
print(f'f-measure:\t{f1_score(y_test, predict)}')
print()
print(classification_report(y_test, predict))
```

### 2.8 Нормализовать данные (StandartScaler или MinMaxScaler).


```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns = X.columns)

X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

### 2.9 Осуществить бинарную классификацию

```python
scaler_model = KNeighborsClassifier(n_neighbors = k)
scaler_model.fit(X_train_scaled, y_train)
predict_scaler = scaler_model.predict(X_test_scaled)
```

### 2.10 Посчитать метрики: accuracy, precision, recall, f-measure, а также составить classification_report

```python
print(f'accurasy:\t{accuracy_score(y_test, predict_scaler)}')
print(f'precision:\t{precision_score(y_test, predict_scaler)}')
print(f'recall:\t\t{recall_score(y_test, predict_scaler)}')
print(f'f-measure:\t{f1_score(y_test, predict_scaler)}')
print()
print(classification_report(y_test, predict_scaler))
```

### 2.11 Сравнить метрики на данных без нормализации и с применением нормализации.

После нормализации данных заметны небольшие улучшения метрик. 

### 2.12 Построить сorrelation heatmap и/или воспользоваться методом corr(). Выбрать наиболее важные признаки эвристически или на основании каких-нибудь вычислений.

```python
import seaborn as sns
```

```python
sns.heatmap(X_scaled.corr(), annot = True)
```

```python
# Вычисляем корреляционную матрицу
corr_matrix = X_scaled.corr()

# Зануляем элементы главной диагонали
for i in range(len(corr_matrix)):
    corr_matrix.iloc[i, i] = 0

# Задаем значение порога для фильтрации
threshold = 0.5

# Фильтруем корреляционную матрицу по порогу
filtered_corr_matrix = corr_matrix.abs() > threshold

# Суммируем значения по каждому признаку для определения важности признаков
feature_importance = filtered_corr_matrix.sum()

# Сортируем признаки по убыванию и получаем список наиболее важных признаков
important_features = feature_importance.sort_values(ascending=False)

print("Наиболее важные признаки:\n", important_features)
```

Наиболее важными параметрами оказались Pregnancies и Age, что и так было видно по корреляционной матрице. 

### 2.13 Осуществить бинарную классификацию с отфильтрованными признаками.

```python
# Выбираем наиболее важные признаки на основе порогового значения
selected_features = ['Pregnancies', 'Age']

# Отфильтровываем данные по выбранным признакам
X_train_scaled_filtered = X_train_scaled[selected_features]
X_test_scaled_filtered = X_test_scaled[selected_features]

# Обучаем модель с отфильтрованными признаками
knn_filtered = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
knn_filtered.fit(X_train_scaled_filtered, y_train)
predict_skaled_filtred = knn_filtered.predict(X_test_scaled_filtered)
```

### 2.14 Посчитать метрики: accuracy, precision, recall, f-measure, а также составить classification_report

```python
print(f'accurasy:\t{accuracy_score(y_test, predict_skaled_filtred)}')
print(f'precision:\t{precision_score(y_test, predict_skaled_filtred)}')
print(f'recall:\t\t{recall_score(y_test, predict_skaled_filtred)}')
print(f'f-measure:\t{f1_score(y_test, predict_skaled_filtred)}')
print()
print(classification_report(y_test, predict_skaled_filtred))
```

### 2.15 Сравнить результаты "до" и "после" фильтрации, сделать выводы.

Метрики однозначно стали хуже. Но в данном случае мы оставили всего два признака из и так не большого набора. В случае, если датасет был бы перегружен признаками - использовать подобный подход было бы логично, но в данном случае это лишь ухудшело итоговый результат.  

### 2.16 Перебрать значения гиперпараметра k в каком-нибудь диапазоне с использованием кросс-валидации на данных

```python
from sklearn.model_selection import cross_val_score
```

```python
train_score = []
test_score = []
crossval_score = []

for k in range(1, 30):
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X_scaled, y, cv = 10)
    print(f"Neighbors: {k}, folds count: 10 avg score: {scores.mean()}")
    crossval_score.append(scores.mean())
    model.fit(X_train_scaled, y_train)
    pred_train = model.predict(X_train_scaled)
    pred_test = model.predict(X_test_scaled)
    train_score.append(accuracy_score(y_train, pred_train))
    test_score.append(accuracy_score(y_test, pred_test))
```

### 2.17 C отфильтрованными признаками. Построить график train/test accuracy в зависимости от значения

```python
from matplotlib import pyplot as plt
plt.plot(range(1,30), train_score, label = 'train_score')
plt.plot(range(1,30), test_score, label = 'test_score')
plt.legend()
plt.grid()
```

## 3.1 Привести описание датасета.

Glass Identification Data Set - датасет содержит 10 признаков (включая Id номер) и классовый атрибут "Type of glass", определяющий тип стекла. | Mg                    | Магний                                                                         | Количественный  |

| K                     | Калий                                                                          | Количественный  |
| Ca                    | Кальций                                                                        | Количественный  |
| Ba                    | Барий                                                                          | Количественный  |

| Type of glass         | Тип стекла (атрибут класса)                                                   | Категориальный  |

### 3.2 Осуществить предобработку данных - "подчистить данные".

```python
X = df_glass.drop('Type of glass', axis = 1)
y = df_glass['Type of glass']
```

### 3.3 Закодировать категориальные признаки при необходимости.

Кодировать данные необходимости нет

### 3.4 Нормализовать данные.

```python
scaler_glass = StandardScaler()
X_glass = scaler_glass.fit_transform(X)
X_glass = pd.DataFrame(X_glass, columns=X.columns)
```

### 3.5 Выбрать какое-нибудь значение гиперпараметра k алгоритма KNN

```python
k = 5
```

### 3.6 Разбить выборку на обучающую и тестовую.

```python
X_glass_train, X_glass_test, y_glass_train, y_glass_test = train_test_split(X_glass, y, test_size=0.33, random_state=42)
```

### 3.7 Осуществить многоклассовую классификацию.

```python
model_glass_KNN = KNeighborsClassifier(n_neighbors=k)
model_glass_KNN.fit(X_glass_train, y_glass_train)
predict_glass_KNN = model_glass_KNN.predict(X_glass_test)
```

### 3.8 Посчитать метрики: accuracy, precision, recall, f-measure, а также составить classification_report

```python
print(f'accurasy:\t{accuracy_score(y_glass_test, predict_glass_KNN)}')
print(f'precision:\t{precision_score(y_glass_test, predict_glass_KNN, average="micro")}')
print(f'recall:\t\t{recall_score(y_glass_test, predict_glass_KNN, average="micro")}')
print(f'f-measure:\t{f1_score(y_glass_test, predict_glass_KNN, average="micro")}')
print()
print(classification_report(y_glass_test, predict_glass_KNN))
```

### 3.9 Построить сorrelation heatmap и/или воспользоваться методом corr(). Выбрать наиболее важные признаки эвристически или на основании каких-нибудь вычислени

```python
sns.heatmap(X_glass.corr(), annot = True, fmt='.1g')
```

```python
# Вычисляем корреляционную матрицу
corr_matrix = X_glass.corr()

# Зануляем элементы главной диагонали
for i in range(len(corr_matrix)):
    corr_matrix.iloc[i, i] = 0

# Задаем значение порога для фильтрации
threshold = 0.50

# Фильтруем корреляционную матрицу по порогу
filtered_corr_matrix = corr_matrix.abs() > threshold

# Суммируем значения по каждому признаку для определения важности признаков
feature_importance = filtered_corr_matrix.sum()

# Сортируем признаки по убыванию и получаем список наиболее важных признаков
important_features = feature_importance.sort_values(ascending=False)

print("Наиболее важные признаки:\n", important_features)
```

### 3.10 Осуществить многоклассовую классификацию с отфильтрованными признаками.

```python
# Выбираем наиболее важные признаки на основе порогового значения
selected_features_glass = ['Id','RI','Mg','Si','Ca']

# Отфильтровываем данные по выбранным признакам
X_train_glass_filtered = X_glass_train[selected_features_glass]
X_test_glass_filtered = X_glass_test[selected_features_glass]

# Обучаем модель с отфильтрованными признаками
knn_glass_filtered = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
knn_glass_filtered.fit(X_train_glass_filtered, y_glass_train)
predict_glass_filtered = knn_glass_filtered.predict(X_test_glass_filtered)
```

### 3.11 Посчитать метрики: accuracy, precision, recall, f-measure, а также составить classification_report

```python
print(f'accurasy:\t{accuracy_score(y_glass_test, predict_glass_filtered)}')
print(f'precision:\t{precision_score(y_glass_test, predict_glass_filtered, average="micro")}')
print(f'recall:\t\t{recall_score(y_glass_test, predict_glass_filtered, average="micro")}')
print(f'f-measure:\t{f1_score(y_glass_test, predict_glass_filtered, average="micro")}')
print()
print(classification_report(y_glass_test, predict_glass_filtered))
```

### 3.12 Сравнить результаты "до" и "после" фильтрации, сделать выводы.

Мтрики снизились, но совсем немного. Но при этом - мы убрали половину признаков. Датасет сам по себе не очень большой, поэтому классификатору не состовляет проблем с ним справиться. Но будь датасет в разы больше - такое уменьшение кол-ва признаков сыграло бы значительную роль в плане производительности, при этом не сильно жертвуя качеством.

### 3.13 Перебрать значения гиперпараметра k в каком-нибудь диапазоне с использованием кросс-валидации на данных с отфильтрованными признаками. Построить график train/test accuracy в зависимости от значения .

```python
train_glass_score = []
test_glass_score = []
crossval_glass_score = []

for k in range(1, 21):
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X_glass, y, cv = 9)
    print(f"Neighbors: {k}, folds count: 10 avg score: {scores.mean()}")
    crossval_glass_score.append(scores.mean())
    model.fit(X_train_glass_filtered, y_glass_train)
    pred_train = model.predict(X_train_glass_filtered)
    pred_test = model.predict(X_test_glass_filtered)
    train_glass_score.append(accuracy_score(y_glass_train, pred_train))
    test_glass_score.append(accuracy_score(y_glass_test, pred_test))
```

```python
plt.plot(range(1, 21), train_glass_score, label = 'train_score')
plt.plot(range(1, 21), test_glass_score, label = 'test_score')
plt.legend()
plt.grid()
```

### 3.14 Для лучшего классификатора привести метрики, classification_report, построить confusion matrix

Лучший результат классификатор достигает при K = 3

```python
model_last = KNeighborsClassifier(n_neighbors=3)
model_last.fit(X_train_glass_filtered, y_glass_train)
predict_last = model_last.predict(X_test_glass_filtered)
```

```python
print(f'accurasy:\t{accuracy_score(y_glass_test, predict_last)}')
print(f'precision:\t{precision_score(y_glass_test, predict_last, average="micro")}')
print(f'recall:\t\t{recall_score(y_glass_test, predict_last, average="micro")}')
print(f'f-measure:\t{f1_score(y_glass_test, predict_last, average="micro")}')
print()
print(classification_report(y_glass_test, predict_last))
```

```python
sns.heatmap(data=confusion_matrix(y_glass_test, predict_last, normalize='true'), square=True, annot=True, fmt='.2f')
```
